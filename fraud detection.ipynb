```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("fraudTest.csv")
```

```python
x=df.drop(columns=["trans_date_trans_time","merchant","category","first","last","gender","street","city","state","job","dob","trans_num","is_fraud"])
```

```python
y=df['is_fraud']
```

# Detect fraud using Logistic Regression machine learning model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.30)
```

```python
model=LogisticRegression(solver='liblinear')
```

```python
model.fit(xtrain,ytrain)
```

```python
model.predict(xtest)
```

```python
logistic_result=model.score(xtest,ytest)
```

```python
logistic_result
```

# Detect  fraud using Decision tree machine learning model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model=DecisionTreeClassifier()
```

```python
tree_model.fit(xtrain,ytrain)
```

```python
tree_result= tree_model.score(xtest,ytest)
```

```python
tree_result
```

# Detect fraud using RandomForest machine learning model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.30, random_state=1)
```

```python
rf= RandomForestClassifier()
```

```python
rf.fit(xtrain, ytrain)
```

```python
rf_result=rf.score(xtest,ytest)
```

```python
rf_result
```

```python
width=0.35
plt.bar(['Logistic Regression'], logistic_result, width, label='Logistic Regression')
plt.bar(['Decision tree'],tree_result,width,label='Decision tree')
plt.bar(['Random Forest'],rf_result,width,label='Random Forest')
plt.xlabel("Machine learning Model",color='red')
plt.ylabel("Result of Each Model", color='brown')
plt.legend(bbox_to_anchor=(1.04, 1.15), ncol=3)
plt.show()
```
